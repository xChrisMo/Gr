```python
import tensorflow as tf
import tensorflow_datasets as tfds
import os
from google.colab import drive
import shutil
import requests
from zipfile import ZipFile
import random
from tqdm import tqdm
```

```python
dataset, info = tfds.load("cassava", as_supervised=True, with_info=True, split=["train", "test", "validation"])
train_ds, test_ds, val_ds = dataset
```

```python
# 2. Mounting drive, makingn paths
drive.mount('/content/drive')
base_dir = "/content/drive/MyDrive/Augmented_Cassava_Dataset"
train_dir = os.path.join(base_dir, "augmented_train")
os.makedirs(train_dir, exist_ok=True)
```

```python
train_ds
```

```python
#Saving w progress bar
train_ds_list = list(train_ds)
with tqdm(total=len(train_ds_list), desc="Saving Cassava images") as pbar:
    for image, label in train_ds_list:
        label_dir = os.path.join(train_dir, str(label.numpy()))
        os.makedirs(label_dir, exist_ok=True)
        img_path = os.path.join(label_dir, f"{tf.timestamp().numpy()}.jpg")
        tf.keras.preprocessing.image.save_img(img_path, image.numpy())
        pbar.update(1)

print("Cassava training data saved to Google Drive.")
```
